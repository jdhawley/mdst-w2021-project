```python
# Extract data from zip
with zipfile.ZipFile('./data.zip', 'r') as zip_ref:
    if not os.path.exists('./data'):
        os.makedirs('./data')
    zip_ref.extractall('./data')
    
# Read in all data
roads = pd.read_csv('./data/raw_road_data.csv')
route_types = pd.read_csv('./data/metadata_route_types.csv')
mtfcc = pd.read_csv('./data/metadata_mtfcc.csv')
```

```python
# Drop outright duplicates
before = len(roads)
roads = roads.drop_duplicates()
after = len(roads)

# Just for kicks and giggles, check how many duplicates were in the data
difference = before - after
print('dropping duplicates eliminated ' + str(difference) + f' rows. ({(difference/before)*100:.2f}%)')
```

```python
# Merge the roads with the metadata
roads = roads.merge(route_types, on='route_type_code', how='left')
roads = roads.merge(mtfcc, on='mtfcc_code', how='left')
```
